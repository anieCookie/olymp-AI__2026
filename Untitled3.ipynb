```python
n_trucks = 3
max_weight = 1500  # кг
max_volume = 10    # м³

items = []
for _, row in df.iterrows():
    items.append({
        'id': int(row['Груз'][-1]),
        'weight': float(row['Масса']),
        'volume': float(row['Объём']),
        'value': int(row['Ценность'])
    })

n_items = len(items)
```

```python
best_value = 0
best_assignments = None

for assignment_num in range(4 ** n_items): 
    assignment = []
    temp = assignment_num
    total_value = 0
    truck_weights = [0, 0, 0]
    truck_volumes = [0, 0, 0]
    
    # Декодируем распределение
    valid = True
    for i in range(n_items):
        truck_idx = temp % 4
        temp //= 4
        
        if truck_idx < 3:  # Груз идет в грузовик
            item = items[i]
            if (truck_weights[truck_idx] + item['weight'] <= max_weight and
                truck_volumes[truck_idx] + item['volume'] <= max_volume):
                truck_weights[truck_idx] += item['weight']
                truck_volumes[truck_idx] += item['volume']
                total_value += item['value']
            else:
                valid = False
                break
    
    if valid and total_value > best_value:
        best_value = total_value
        best_assignments = (assignment_num, truck_weights, truck_volumes)

print(f"Максимальная суммарная ценность: {best_value}")
```

```python
from itertools import product

n = len(items)
best = 0

for assignment in product(range(4), repeat=n):
    trucks = [{ 'weight': 0, 'volume': 0, 'value': 0 } for _ in range(3)]
    valid = True
    total_value = 0
    
    for i, truck_idx in enumerate(assignment):
        if truck_idx > 0: 
            truck = trucks[truck_idx-1]
            item = items[i]
            
            if (truck['weight'] + item['weight'] <= 1500 and
                truck['volume'] + item['volume'] <= 10):
                truck['weight'] += item['weight']
                truck['volume'] += item['volume']
                truck['value'] += item['value']
                total_value += item['value']
            else:
                valid = False
                break
    
    if valid and total_value > best:
        best = total_value

print(f"Ответ: {best}")
```

```python
# 10
'''Укажите пару товаров, которая встречается в наибольшем числе потребительских корзин'''

from itertools import combinations
from collections import Counter

# Все пары товаров в одну строку
all_pairs = [pair for basket in basket_df['product_name']
             if len(basket) >= 2
             for pair in combinations(sorted(basket), 2)]

most_common_pair = Counter(all_pairs).most_common(1)[0]

print(f"Самая частая пара: {most_common_pair[0]}")
print(f"Количество корзин: {most_common_pair[1]}")
```

```python
# 11
'''Укажите “client_id” клиента, у которого потребительская корзина имеет наибольшее пересечение с потребительской корзиной клиента с client_id = 1011'''

# Находим потребительскую корзину клиента 1011
target_basket = set(basket_df[basket_df['client_id'] == 1011]['product_name'].iloc[0])

# Считаем размер пересечения с корзиной каждого другого клиента
intersections = []
for _, row in basket_df[basket_df['client_id'] != 1011].iterrows():
    client_id = row['client_id']
    basket = set(row['product_name'])
    intersection_size = len(target_basket & basket)  # Размер пересечения
    intersections.append((client_id, intersection_size))

# Находим клиента с максимальным пересечением
most_similar_client = max(intersections, key=lambda x: x[1])[0]

print(f"Клиент с наибольшим пересечением корзины с клиентом 1011: {most_similar_client}")
```

```python
# 4

df['date_time'] = pd.to_datetime(df['date_time'])
df['date'] = df['date_time'].dt.date

# Создаем DataFrame с ежедневными конверсиями
def calculate_daily_conversions(df):
    dates = sorted(df['date'].unique())
    results = []
    
    for date in dates:
        day_data = df[df['date'] == date]
        
        # Уникальные пользователи на каждом шаге за день
        visit_users = day_data[day_data['action'] == 'visit']['user_id'].nunique()
        click_users = day_data[day_data['action'] == 'click']['user_id'].nunique()
        cart_users = day_data[day_data['action'] == 'cart']['user_id'].nunique()
        purchase_users = day_data[day_data['action'] == 'purchase']['user_id'].nunique()
        
        # Расчет конверсий
        conversion_visit_click = click_users / visit_users if visit_users > 0 else 0
        conversion_visit_cart = cart_users / visit_users if visit_users > 0 else 0
        conversion_visit_purchase = purchase_users / visit_users if visit_users > 0 else 0
        
        results.append({
            'date': date,
            'visit_to_click': conversion_visit_click,
            'visit_to_cart': conversion_visit_cart,
            'visit_to_purchase': conversion_visit_purchase,
            'total_visits': visit_users
        })
    
    return pd.DataFrame(results)

# Расчет данных
daily_conversions = calculate_daily_conversions(df)

# Построение общего графика
plt.figure(figsize=(12, 6))

# Три линии на одном графике без маркеров
plt.plot(daily_conversions['date'], daily_conversions['visit_to_click'], 
         linewidth=2.5, label='Визит → Клик', color='#1f77b4')
plt.plot(daily_conversions['date'], daily_conversions['visit_to_cart'], 
         linewidth=2.5, label='Визит → Корзина', color='#ff7f0e')
plt.plot(daily_conversions['date'], daily_conversions['visit_to_purchase'], 
         linewidth=2.5, label='Визит → Покупка', color='#2ca02c')

# Настройка внешнего вида
plt.title('Динамика конверсий по дням', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Уровень конверсии', fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)

# Форматирование оси Y в процентах
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.1%}'))

# Поворот дат для лучшей читаемости
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Вывод статистики по конверсиям
print("СРЕДНИЕ ЗНАЧЕНИЯ КОНВЕРСИЙ ЗА ВЕСЬ ПЕРИОД:")
print(f"Визит → Клик: {daily_conversions['visit_to_click'].mean():.3%}")
print(f"Визит → Корзина: {daily_conversions['visit_to_cart'].mean():.3%}")
print(f"Визит → Покупка: {daily_conversions['visit_to_purchase'].mean():.3%}")
```

```python
# 2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def find_outliers_iqr(column):
    """
    Находит выбросы с помощью метода межквартильного размаха (IQR)
    """
    # Рассчитываем квартили
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    
    # Границы для выбросов
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Находим выбросы
    outliers = column[(column < lower_bound) | (column > upper_bound)]
    
    return outliers, lower_bound, upper_bound, Q1, Q3, IQR

# Применяем к вашему столбцу
outliers, lower_bound, upper_bound, Q1, Q3, IQR = find_outliers_iqr(A['orders_cnt_per_user'])

print("=== АНАЛИЗ ВЫБРОСОВ МЕТОДОМ IQR ===")
print(f"Q1 (25-й перцентиль): {Q1:.2f}")
print(f"Q3 (75-й перцентиль): {Q3:.2f}")
print(f"IQR (межквартильный размах): {IQR:.2f}")
print(f"Нижняя граница: {lower_bound:.2f}")
print(f"Верхняя граница: {upper_bound:.2f}")
print(f"Найдено выбросов: {len(outliers)}")
print(f"Выбросы: {outliers.tolist()}")
```
