### DBSCAN

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(
    eps=0.5,        # радиус соседства
    min_samples=5,  # минимальное число точек для кластера
    metric="euclidean"
)

dbscan.fit(X)
labels = dbscan.labels_  # -1 → шум
```
